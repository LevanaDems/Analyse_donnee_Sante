### Visualisation des dispertion et detection des outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pr√©parer les colonnes num√©riques (hors target)
sns.set(style="whitegrid")
numeric_cols = df.select_dtypes(include=["float64", "int64"]).columns.tolist()
numeric_cols.remove("TenYearCHD")

```

```python

#Nous allons d√©tecter les outliers en utilisant IQR


import numpy as np

# d√©finisons une fonction
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return len(outliers), lower_bound, upper_bound
```

```python
check_column = ['cigsPerDay', 'totChol', 'BMI', 'heartRate', 'glucose', 'sysBP', 'diaBP']
outliers_summary = []

for column in check_column:
    count, lower, upper = detect_outliers(df, column)
    outliers_summary.append({
        "variable": column,
        "Nombre d'outliers": count,
        "Seuil bas": lower,
        "Seuil haut": upper
    })

outliers_df = pd.DataFrame(outliers_summary)
outliers_df
```

```python
# Nous allons passer au traitement des outliers

# Appliquer une imputation douce des outliers par la borne sup√©rieure

# Dictionnaire des seuils sup√©rieurs pour remplacement
outlier_caps = {
    'glucose': 104.5,
    'totChol': 346.0,
    'BMI': 35.47,
    'heartRate': 105.5
}

# Remplacer les valeurs sup√©rieures √† la borne par la borne elle-m√™me
for col, cap in outlier_caps.items():
    df[col] = np.where(df[col] > cap, cap, df[col])

# V√©rification rapide des nouvelles valeurs max
new_max_values = df[list(outlier_caps.keys())].max()
new_max_values
```

# **MODELISATION**

## PREPARATION DES DONNEES

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# S√©paration features / target
X = df.drop(columns=["TenYearCHD"])
y = df["TenYearCHD"]

# Split en conservant le d√©s√©quilibre (stratify)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Standardisation des variables num√©riques
numeric_cols = X.select_dtypes(include=["float64", "int64"]).columns
scaler = StandardScaler()
X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols] = scaler.transform(X_test[numeric_cols])

# V√©rification des classes
print("R√©partition dans y_train :\n", y_train.value_counts(normalize=True))
print("R√©partition dans y_test :\n", y_test.value_counts(normalize=True))
print("X_train shape :", X_train.shape)
```

ENTRAINEMENT

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Regression logistique avec class_weight
log_model = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
log_model.fit(X_train, y_train)

# Random Forest avec class_weight
rf_model = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
rf_model.fit(X_train, y_train)

# Pr√©dictions
y_pred_log = log_model.predict(X_test)
y_prob_log = log_model.predict_proba(X_test)[:, 1]

y_pred_rf = rf_model.predict(X_test)
y_prob_rf = rf_model.predict_proba(X_test)[:, 1]

# √âvaluation
print("\n---  REGRESSION LOGISTIQUE ---")
print("ROC-AUC:", roc_auc_score(y_test, y_prob_log))
print(classification_report(y_test, y_pred_log))
print("Matrice de confusion :\n", confusion_matrix(y_test, y_pred_log))

print("\n--- RANDOM FOREST ---")
print("ROC-AUC:", roc_auc_score(y_test, y_prob_rf))
print(classification_report(y_test, y_pred_rf))
print("Matrice de confusion :\n", confusion_matrix(y_test, y_pred_rf))

# Courbe ROC
fpr_log, tpr_log, _ = roc_curve(y_test, y_prob_log)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)

plt.figure(figsize=(8, 5))
plt.plot(fpr_log, tpr_log, label=' Regression Logistique (AUC = {:.2f})'.format(roc_auc_score(y_test, y_prob_log)))
plt.plot(fpr_rf, tpr_rf, label='Random Forest (AUC = {:.2f})'.format(roc_auc_score(y_test, y_prob_rf)))
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("Taux de faux positifs")
plt.ylabel("Taux de vrais positifs")
plt.title("Courbe ROC - Mod√®les sur donn√©es d√©s√©quilibr√©es")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

Matrice de confusion

On constate que enutilisant toutes les variables et en conservant le d√©s√©quilibre la regression logistique semble mieux pr√©dire (avec recall= 58% et un accuracy = 67%)les donn√©es que le random forest.

```python
import seaborn as sns

# Fonction pour afficher une matrice de confusion
def plot_confusion_matrix(cm, title):
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['Non √† risque (0)', '√Ä risque (1)'],
                yticklabels=['Non √† risque (0)', '√Ä risque (1)'])
    plt.xlabel("Pr√©diction")
    plt.ylabel("Vrai")
    plt.title(title)
    plt.tight_layout()
    plt.show()

# üî≤ Matrices de confusion
cm_log = confusion_matrix(y_test, y_pred_log)
cm_rf = confusion_matrix(y_test, y_pred_rf)

# üîç Affichage graphique
plot_confusion_matrix(cm_log, "Matrice de confusion - Logistic Regression")
plot_confusion_matrix(cm_rf, "Matrice de confusion - Random Forest")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les donn√©es et s√©lectionner les variables pertinentes
df = pd.read_csv("maladies_cardiaques.csv")

features = [
    'age', 'sysBP', 'diaBP', 'BMI', 'glucose', 'totChol',
    'prevalentHyp', 'diabetes', 'BPMeds', 'male'
]

df = df[features + ['TenYearCHD']]

# Imputer les valeurs manquantes
from sklearn.impute import SimpleImputer
median_imputer = SimpleImputer(strategy='median')
df[['BMI', 'glucose', 'totChol']] = median_imputer.fit_transform(df[['BMI', 'glucose', 'totChol']])
mode_imputer = SimpleImputer(strategy='most_frequent')
df[['BPMeds']] = mode_imputer.fit_transform(df[['BPMeds']])

# Winsorisation douce
winsor_caps = {
    'glucose': 104.5,
    'totChol': 346.0,
    'BMI': 35.47
}
for col, cap in winsor_caps.items():
    df[col] = np.where(df[col] > cap, cap, df[col])

# S√©paration X/y
X = df.drop(columns=["TenYearCHD"])
y = df["TenYearCHD"]

#  Split (sans SMOTE)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Standardisation des colonnes num√©riques
numeric_cols = ['age', 'sysBP', 'diaBP', 'BMI', 'glucose', 'totChol']
scaler = StandardScaler()
X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols] = scaler.transform(X_test[numeric_cols])

#  Mod√®le :  Regression Logistique
model = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
model.fit(X_train, y_train)

#  √âvaluation
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]

print("\n--- Regression logistique (Variables pertinentes, sans SMOTE) ---")
print("ROC-AUC :", roc_auc_score(y_test, y_prob))
print(classification_report(y_test, y_pred))
print("Matrice de confusion :\n", confusion_matrix(y_test, y_pred))

# Matrice de confusion
def plot_confusion_matrix(cm, title):
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['Non √† risque (0)', '√Ä risque (1)'],
                yticklabels=['Non √† risque (0)', '√Ä risque (1)'])
    plt.xlabel("Pr√©dit")
    plt.ylabel("Vrai")
    plt.title(title)
    plt.tight_layout()
    plt.show()

plot_confusion_matrix(confusion_matrix(y_test, y_pred), "Matrice de confusion - Mod√®le final")
```

En se basant sur les variables pertinantes on peut voir que le recall de la regression lin√©aire est de 60% ce qui est nettement mieux que le pr√©c√©dent.
